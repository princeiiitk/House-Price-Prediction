```python
X=df.drop('price',axis=1);
y=df['price']
df['furnishingstatus'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
le=LabelEncoder();
oe=OrdinalEncoder(categories=[['unfurnished','semi-furnished','furnished']]);

categorical_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'furnishingstatus','prefarea']

for col in categorical_cols:
    if df[col].dtype == 'object':
        X[col] = le.fit_transform(X[col])
    else:
        pass



df.info()

```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=284);
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)


print(f"Maximum R-squared score: {r2*100}")
```
